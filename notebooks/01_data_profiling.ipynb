```python
# ============================================
# SENTINEL AI - Data Profiling
# ============================================

import pandas as pd

df_train_transaction = pd.read_csv('train_transaction.csv')
df_train_identity = pd.read_csv('train_identity.csv')

# --------------------------------------------
# Section 1 - Row and Column Counts
# --------------------------------------------
print("Train Transaction shape:", df_train_transaction.shape)
print("Train Identity shape:", df_train_identity.shape)

# --------------------------------------------
# Section 2 - Column Names
# --------------------------------------------
print("\nTransaction columns:\n", df_train_transaction.columns.tolist())
print("\nIdentity columns:\n", df_train_identity.columns.tolist())

# --------------------------------------------
# Section 3 - Data Types
# --------------------------------------------
print("\nTrain Transaction DTypes:\n", df_train_transaction.dtypes)
print("\nTrain Identity DTypes:\n", df_train_identity.dtypes)

# --------------------------------------------
# Section 4 - Primary Key Validation (TransactionID)
# --------------------------------------------
print("\nTransaction table - TransactionID nulls:", df_train_transaction['TransactionID'].isnull().sum())
print("Transaction table - TransactionID unique:", df_train_transaction['TransactionID'].is_unique)

print("\nIdentity table - TransactionID nulls:", df_train_identity['TransactionID'].isnull().sum())
print("Identity table - TransactionID unique:", df_train_identity['TransactionID'].is_unique)

# --------------------------------------------
# Section 5 - Duplicate Check (explicit count)
# --------------------------------------------
print("\nDuplicate TransactionIDs in transaction table:", df_train_transaction['TransactionID'].duplicated().sum())
print("Duplicate TransactionIDs in identity table:", df_train_identity['TransactionID'].duplicated().sum())

# --------------------------------------------
# Section 6 - Join Validation
# --------------------------------------------
matched_ids = df_train_transaction['TransactionID'].isin(df_train_identity['TransactionID']).sum()
print("\nMatched TransactionIDs:", matched_ids)
print("Unmatched:", 590540 - matched_ids)

# --------------------------------------------
# Section 7 - Missing Value Profile (sorted, full)
# --------------------------------------------
print("\nTrain Transaction - Null Percentage:\n")
print((df_train_transaction.isnull().sum()/len(df_train_transaction)*100).sort_values(ascending=False))

print("\nTrain Identity - Null Percentage:\n")
print((df_train_identity.isnull().sum()/len(df_train_identity)*100).sort_values(ascending=False))

# --------------------------------------------
# Section 8 - Business Rule Validation
# --------------------------------------------
invalid_amt = df_train_transaction[df_train_transaction['TransactionAmt'] <= 0]
print("\nRows with TransactionAmt <= 0:", len(invalid_amt))
print("Min TransactionAmt:", df_train_transaction['TransactionAmt'].min())
print("Max TransactionAmt:", df_train_transaction['TransactionAmt'].max())

print("\nisFraud value counts:\n", df_train_transaction[['isFraud']].value_counts())
print("isFraud nulls:", df_train_transaction['isFraud'].isnull().sum())

# --------------------------------------------
# Section 9 - D-columns Non-negative Check
# --------------------------------------------
d_cols = df_train_transaction.filter(regex=r'^D\d+$')
print("\nMinimum values of D1 to D15 columns:\n", d_cols.min())

# --------------------------------------------
# Section 10 - Statistical Summary
# --------------------------------------------
print("\nTrain Transaction Describe:\n", df_train_transaction.describe(include='all'))
print("\nTrain Identity Describe:\n", df_train_identity.describe(include='all'))
```

```python
# Columns with 90%+ nulls
txn_null_pct = (df_train_transaction.isnull().sum()/len(df_train_transaction)*100)
print("Transaction columns with 90%+ nulls:\n", txn_null_pct[txn_null_pct >= 90].sort_values(ascending=False))

id_null_pct = (df_train_identity.isnull().sum()/len(df_train_identity)*100)
print("\nIdentity columns with 90%+ nulls:\n", id_null_pct[id_null_pct >= 90].sort_values(ascending=False))
```
